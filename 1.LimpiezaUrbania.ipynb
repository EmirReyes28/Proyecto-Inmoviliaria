# Limpieza de datos

1. Asegurarse de que no halla filas vacias o duplicadas.
2. La columna **precio** contiene el valor tanto en soles como en dolares, o solo el valor en soles.Se debe separar los valores en columnas correspondientes.
3. La columna **metraje** y **dormitorios** contienen valores que pertenecen a un rango. Esto representa a departamentos en venta en plano, lo cual no contempla nuestro caso de estudio.

```python
df2 = df2.dropna(how='all')
```

```python
display(df2.head(20))
```

```python
df1['precio'] = df1['precio'].astype(str)
df2['precio'] = df2['precio'].astype(str)
```

```python
# Función para limpiar y separar la columna precio, y hacer la conversión de moneda
def clean_and_split_price(df, exchange_rate=3.7):
    # Limpiar la columna precio y separar en soles y dólares
    df['Precio Soles'] = df['precio'].apply(lambda x: x.split(' · ')[0] if 'S/' in x else None)
    df['Precio Dolares'] = df['precio'].apply(lambda x: x.split(' · ')[1] if ' · ' in x else x if 'USD' in x else None)
    
    # Eliminar caracteres no numéricos y convertir a float
    df['Precio Soles'] = df['Precio Soles'].str.replace(r'[^\d.]', '', regex=True).astype(float)
    df['Precio Dolares'] = df['Precio Dolares'].str.replace(r'[^\d.]', '', regex=True).astype(float)
    
    # Convertir valores faltantes
    df['Precio Soles'] = df.apply(lambda row: row['Precio Dolares'] * exchange_rate 
                                  if pd.isna(row['Precio Soles']) and not pd.isna(row['Precio Dolares']) else row['Precio Soles'], 
                                  axis=1)
    df['Precio Dolares'] = df.apply(lambda row: row['Precio Soles'] / exchange_rate 
                                  if pd.isna(row['Precio Dolares']) and not pd.isna(row['Precio Soles']) else row['Precio Dolares'], 
                                  axis=1)
    
    # Formatear los valores numéricos para evitar notación científica
    df['Precio Soles'] = df['Precio Soles'].map('{:,.2f}'.format)
    df['Precio Dolares'] = df['Precio Dolares'].map('{:,.2f}'.format)
    
    # Eliminar la columna original
    df = df.drop(columns=['precio'])
    
    return df
```

```python
df1 = clean_and_split_price(df1)
df2 = clean_and_split_price(df2)
```

```python
display(df1.head(20))
display(df2.head(20))
```

```python
df_combined = pd.concat([df1, df2], ignore_index=True)
```

```python
df_combined = df_combined.drop_duplicates().dropna()
```
