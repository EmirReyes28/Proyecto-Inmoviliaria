# Limpieza de datos

1. Asegurarse de que no halla filas vacias o duplicadas.
2. Las columnas **metraje** y **dormitorios** contienen valores que pertenecen a un rango. Esto representa a departamentos en venta en plano, lo cual no contempla nuestro caso de estudio.
3. La columna **metraje** es una cadena de valores numericos como strings. Solo necesitamos los valores numericos.
3. La columna **precio** contiene el valor tanto en soles como en dolares, o solo el valor en soles.Se debe separar los valores en columnas correspondientes.

### 1. Filas vacias

Los dataframes cuentan con un total de 5 columnas, por tal motivo si en una fila existen mas de 3 valores NaN quiere decir que esa fila no me sirve como valor de estudio.

```python
filas_nan1 = df1.isnull().sum(axis=1).eq(5)
print(f"El número de filas totalmente vacias con valores NaN es: {filas_nan1.sum()}")
filas_nan2 = df2.isnull().sum(axis=1).eq(5)
print(f"El número de filas totalmente vacias con valores NaN es: {filas_nan2.sum()}")
```

Sobreescribimos los dataframe con valores que no tengan NaN entre sus filas

```python
df1=df1[~filas_nan1]
df2=df2[~filas_nan2]
```

```python
display(df1)
display(df2)
```

### 2. Filtrado de departamentos en venta en plano

Existen publicaciones que promocionan la venta de departamentos en plano y no tienen un valor especifico.Esto afectaria la integridad del dataframe por tal motivo se decidio eliminarlas.

```python
df1 = df1[~df1['metraje'].str.contains('a', na=False) & ~df1['dormitorios'].str.contains('a', na=False)]
df2 = df2[~df2['metraje'].str.contains('a', na=False) & ~df2['dormitorios'].str.contains('a', na=False)]
```

```python
display(df1)
display(df2)
```

### 3. Limpieza de la columna "metraje" y "dormitorios" solo valores numericos

```python
def extraer_numeros(valor):
    if pd.isna(valor):
        return 0
    # Usar expresión regular para extraer números
    numeros = re.findall(r'\d+', str(valor))
    # Si se encuentra algún número, convertir a entero, si no, devolver 0
    return int(numeros[0]) if numeros else 0
```

```python
df1['metraje'] = df1['metraje'].astype(str).apply(extraer_numeros)
df2['metraje'] = df2['metraje'].astype(str).apply(extraer_numeros)
```

```python
df1['dormitorios'] = df1['dormitorios'].astype(str).apply(extraer_numeros)
df2['dormitorios'] = df2['dormitorios'].astype(str).apply(extraer_numeros)
```

```python
display(df1)
display(df2)
```

### 4. Limpieza de la columna "precio"

En la columna **precio** algunas filas tienen su valor tanto en soles o dolares o ambos pero como cadena de texto. 

```python
df1['precio'] = df1['precio'].astype(str)
df2['precio'] = df2['precio'].astype(str)
```

```python
# Función para limpiar y separar la columna precio, y hacer la conversión de moneda
def clean_and_split_price(df, exchange_rate=3.7):
    # Limpiar la columna precio y separar en soles y dólares
    df['Precio Soles'] = df['precio'].apply(lambda x: x.split(' · ')[0] if 'S/' in x else None)
    df['Precio Dolares'] = df['precio'].apply(lambda x: x.split(' · ')[1] if ' · ' in x else x if 'USD' in x else None)

    # Eliminar caracteres no numéricos y convertir a float
    df['Precio Soles'] = df['Precio Soles'].str.replace(r'[^\d.]', '', regex=True).astype(float)
    df['Precio Dolares'] = df['Precio Dolares'].str.replace(r'[^\d.]', '', regex=True).astype(float)

    # Convertir valores faltantes
    df['Precio Soles'] = df.apply(lambda row: row['Precio Dolares'] * exchange_rate 
                                  if pd.isna(row['Precio Soles']) and not pd.isna(row['Precio Dolares']) else row['Precio Soles'], 
                                  axis=1)
    df['Precio Dolares'] = df.apply(lambda row: row['Precio Soles'] / exchange_rate 
                                  if pd.isna(row['Precio Dolares']) and not pd.isna(row['Precio Soles']) else row['Precio Dolares'], 
                                  axis=1)

    # Formatear los valores numéricos para evitar notación científica
    df['Precio Soles'] = df['Precio Soles'].map('{:,.2f}'.format)
    df['Precio Dolares'] = df['Precio Dolares'].map('{:,.2f}'.format)

    # Eliminar la columna original
    df = df.drop(columns=['precio'])

    return df
```

```python
df1=clean_and_split_price(df1)
df2=clean_and_split_price(df2)
```

```python
display(df1)
display(df2)
```

## Union de los Dataframes limpios

```python
df3 = pd.concat([df1, df2], ignore_index=True)
```
