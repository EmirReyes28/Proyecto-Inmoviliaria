# inicio del codigo

```python
df1['precio'] = df1['precio'].astype(str)
df2['precio'] = df2['precio'].astype(str)
```

```python
# Función para limpiar y separar la columna precio
def clean_and_split_price(df):
    # Limpiar la columna precio y separar en soles y dólares
    df['Precio Soles'] = df['precio'].apply(lambda x: x.split(' · ')[0] if 'S/' in x else None)
    df['Precio Dolares'] = df['precio'].apply(lambda x: x.split(' · ')[1] if ' · ' in x else x if 'USD' in x else None)
    
    # Eliminar caracteres no numéricos y convertir a float
    df['Precio Soles'] = df['Precio Soles'].str.replace(r'[^\d.]', '', regex=True).astype(float)
    df['Precio Dolares'] = df['Precio Dolares'].str.replace(r'[^\d.]', '', regex=True).astype(float)
    
    # Eliminar la columna original
    df = df.drop(columns=['precio'])
    
    return df
```

```python
df1 = clean_and_split_price(df1)
df2 = clean_and_split_price(df2)
```
