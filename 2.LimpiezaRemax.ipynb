## Normalizado del nombre de las columnas

```python
df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]
```

## Limpieza de datos

En este apartado se obtendra los valores netos sean INT o FLOAT de las diferentes columnas

```python
df['precio_soles'] = pd.to_numeric(df['precio_soles'], errors='coerce')
df['precio_dolares'] = pd.to_numeric(df['precio_dolares'], errors='coerce')
df['area_construida'] = df['area_construida'].str.extract('(\d+\.?\d*)').astype(float)
df['area_total'] = df['area_total'].str.extract('(\d+\.?\d*)').astype(float)
df['pisos'] = df['pisos'].str.extract('(\d+)').astype(int)
df['dormitorios'] = df['dormitorios'].str.extract('(\d+)').astype(int)
df['baños'] = df['baños'].str.extract('(\d+)').astype(int)
df['cochera'] = df['cochera'].str.extract('(\d+)').astype(int)
```

```python
display(df.head(20))
display(df.shape)
```

## Filtrado de informacion por Departamento o Casa

El Dataframe creado a partir del archivo csv contiene informacion de de casas,departamentos,terrenos y chacras. En el caso de estudio solo requerimos informacion relacionada a casas y departamentos.

```python
filtro1 = df['tipo_de_inmueble'].str.contains('DEPARTAMENTO|CASA', case=False, na=False)
df_filtrado = df[filtro1]
```

```python
display(df_filtrado.head(20))
display(df_filtrado.shape)
```
