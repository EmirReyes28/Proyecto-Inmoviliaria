# inicio del codigo

```python
df1['precio'] = df1['precio'].astype(str)
```

```python
def extraer_precios(precio):
    precio_soles = None
    precio_dolares = None
    
    if 'S/' in precio:
        try:
            precio_soles = int(precio.split('S/')[1].split('·')[0].strip().replace(',', ''))
        except ValueError:
            pass
    if 'USD' in precio:
        try:
            precio_dolares = int(precio.split('USD')[1].strip().replace(',', ''))
        except ValueError:
            pass
    
    return pd.Series([precio_soles, precio_dolares])
```

```python
# Aplicar la función al dataframe
df1[['precio soles(S/)', 'precio dolares(USD)']] = df1['precio'].apply(extraer_precios)
```

```python
df1.drop(columns=['precio'])
```

```python
# Tasa de cambio de soles a dólares
tasa_cambio = 3.70

# Supongamos que tienes un DataFrame llamado 'df' y quieres convertir los valores en soles a dólares
df1['precio dolares(USD)'] = df1.apply(lambda row: '{:.2f}'.format(float(row['precio soles(S/)']) / tasa_cambio) 
                                     if pd.isna(row['precio dolares(USD)']) else row['precio dolares(USD)'], axis=1)

```
