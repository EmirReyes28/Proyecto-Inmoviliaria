# inicio del codigo

```python
df['precio'] = df['precio'].astype(str)
```

```python
def extraer_precios(precio):
    precio_soles = None
    precio_dolares = None
    
    if 'S/' in precio:
        try:
            precio_soles = int(precio.split('S/')[1].split('·')[0].strip().replace(',', ''))
        except ValueError:
            pass
    if 'USD' in precio:
        try:
            precio_dolares = int(precio.split('USD')[1].strip().replace(',', ''))
        except ValueError:
            pass
    
    return pd.Series([precio_soles, precio_dolares])
```

```python
# Aplicar la función al dataframe
df[['precio soles(S/)', 'precio dolares(USD)']] = df['precio'].apply(extraer_precios)
```

```python
df.head(20)
```

```python
# Tasa de cambio de soles a dólares
tasa_cambio = 3.70

# Convertir los valores en soles a dólares donde no hay valor en dólares y formatear los valores en dólares
df['precio dolares(USD)'] = df.apply(lambda row: '{:.2f}'.format(row['precio soles(S/)'] / tasa_cambio) if pd.isna(row['precio dolares(USD)']) else '{:.2f}'.format(row['precio dolares(USD)']), axis=1)

```

```python
df = df.drop(columns=['precio'])
```
